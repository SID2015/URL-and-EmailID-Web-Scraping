```python
#Importing requried libraries

import urllib.request

from bs4 import BeautifulSoup

urlToScrape = "Your URL that you want to scrape"


page = urllib.request.urlopen(urlToScrape)

soup = BeautifulSoup(page, features='html.parser')

print('\n\n !!!  /// Page Scraped ///  !!! \n\n')

print('\n\n !!! Title of the page !!! \n\n')

print(soup.title.string)

print('\n\n  !! All the URLs in the Web Page !! \n\n')

all_links = soup.find_all('a')

print('\n\n !! Total Number of URLs present !! \n\n = ' , len(all_links) )


print('\n\n Last 5 URLs in the page are \n\n ' )

if len(all_links) > 5 :
    last_5 = all_links[len(all_links) - 5 : ]
    
    for url in last_5:
        print(url.get('href'))

        
emails = []  

for url in all_links:
    if(str(url.get('href')).find('@') > 0):
        emails.append(url.get('href'))
        
        

print('\n\n !! Some of the emails are!! \n\n')

for email in emails[:5]:
    print(email)

```
